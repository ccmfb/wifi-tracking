```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.tri as tri
```

```python

# Generate irregularly spaced data points
x = np.random.rand(50) * 4 - 2
y = np.random.rand(50) * 4 - 2

x = [
    0, 1, 1, 2, 2, 0
]
y = [
    2, 2, 1, 1, 0, 0
]
z = np.sin(x) + np.cos(y)

# Create a triangular grid
triang = tri.Triangulation(x, y)

# Plot the triangulation
plt.figure()
plt.triplot(triang, 'bo-', lw=1)  # Plot the triangulation
# plt.tricontourf(triang, z)  # Plot the contour
plt.title('Basic Triangulation')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
